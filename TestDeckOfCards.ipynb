```python
import Deck
```

```python
# Make a deck of cards
deck = Deck.Deck()
print("There are {} cards in the deck".format(len(deck)))
print(deck)
```

```python
# Make deck with 2 standard deck of cards
deck = Deck.Deck(2)
print("There are {} cards in the deck".format(len(deck)))
print(deck)
```

```python
#  Add 6 more decks of cards to deck for a 8 deck shoe
deck.AddDecks(6)
print("There are {} cards in the deck".format(len(deck)))
```

```python
# Make a new single deck of cards and shuffle them
deck = Deck.Deck()
print("Deck before shuffling:\n{}".format(deck))
deck.Shuffle()
print("Deck after shuffling:\n{}".format(deck))
```

```python
# Deal entire deck one at a time
deck = Deck.Deck()
for i in range(len(deck)):
    card = deck.DealOneCard()
    print("Card dealt was {}".format(card))
    print("\tThere are {} cards left in the deck".format(len(deck)))
```
